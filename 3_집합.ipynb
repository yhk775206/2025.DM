## 1. Í∞ÑÎã®Ìïú ÏßëÌï© Ïó∞ÏÇ∞

```python
s1=set([1,2,3])  # ÏßëÌï© {1,2,3}ÏÑ†Ïñ∏
print(s1)  # {1, 2, 3}
s2=set([2,3,4])
print(s2)

print(s1 & s2)  # ÍµêÏßëÌï© {2, 3}
print(s1 | s2)  # Ìï©ÏßëÌï© {1,2,3,4}
print(s1 - s2)  # Ï∞®ÏßëÌï© {1}
```

## 2. Îëê ÏùÄÌñâÏùò Í≥†Í∞ù ÏßëÌï© Ìï©Î≥ë

```python
# Îëê ÏùÄÌñâÏùò Í≥†Í∞ù ÏßëÌï© ÏÑ†Ïñ∏
bankA_customers = set(["ÍπÄÏ≤†Ïàò", "Ïù¥ÏòÅÌù¨", "Î∞ïÎØºÏàò", "ÏµúÏßÄÏö∞"])
bankB_customers = set(["Ïù¥ÏòÅÌù¨", "ÏµúÏßÄÏö∞", "Ï†ïÏö∞ÏÑ±", "ÌïúÏßÄÎØº"])

print("ÏùÄÌñâ A Í≥†Í∞ù:", bankA_customers)
print("ÏùÄÌñâ B Í≥†Í∞ù:", bankB_customers)

# Îëê ÏùÄÌñâÏùò Ï†ÑÏ≤¥ Í≥†Í∞ù ÏßëÌï© (Ìï©ÏßëÌï©)
total_customers = bankA_customers | bankB_customers
print("Ìï©Î≥ë ÌõÑ Ï†ÑÏ≤¥ Í≥†Í∞ù:", total_customers)

# Îëê ÏùÄÌñâ Î™®ÎëêÏóê ÏûàÎäî Í≥µÌÜµ Í≥†Í∞ù (ÍµêÏßëÌï©)
common_customers = bankA_customers & bankB_customers
print("Í≥µÌÜµ Í≥†Í∞ù:", common_customers)

# A ÏùÄÌñâÏóêÎßå ÏûàÎäî Í≥†Í∞ù (Ï∞®ÏßëÌï©)
only_A_customers = bankA_customers - bankB_customers
print("A ÏùÄÌñâÏóêÎßå ÏûàÎäî Í≥†Í∞ù:", only_A_customers)

# B ÏùÄÌñâÏóêÎßå ÏûàÎäî Í≥†Í∞ù (Ï∞®ÏßëÌï©)
only_B_customers = bankB_customers - bankA_customers
print("B ÏùÄÌñâÏóêÎßå ÏûàÎäî Í≥†Í∞ù:", only_B_customers)
```

### 3. ÏûÑÏùòÏùò ÏûÖÎ†• ÏÉâÏÉÅÏùÑ Î∞õÏïÑ, 12Í∞ÄÏßÄ ÌååÎ≤ÑÏπ¥Ïä§ÌÖî ÏÉâÏó∞ÌïÑ ÏÉâÏÉÅ ÏßëÌï©ÏóêÏÑú Í∞ÄÏû• Í∞ÄÍπåÏö¥ ÏÉâÏÉÅÏúºÎ°ú Îß§Ìïë

```python
import numpy as np

# 12Í∞ÄÏßÄ ÌååÎ≤ÑÏπ¥Ïä§ÌÖî ÏÉâÏó∞ÌïÑ ÏÉâÏÉÅ ÏßëÌï© (RGB Í∞í ÏòàÏãú)
faber_castell_colors = {
    "White": (255, 255, 255),
    "Black": (0, 0, 0),
    "Brown": (150, 75, 0),
    "Orange": (255, 127, 0),
    "Red": (255, 0, 0),
    "Yellow": (255, 255, 0),
    "Light Green": (0, 255, 0),
    "Green": (0, 128, 0),
    "Light Blue": (0, 255, 255),
    "Blue": (0, 0, 255),
    "Purple": (128, 0, 128),
    "Pink": (255, 192, 203)
}

# ÏûÖÎ†• ÏÉâÏÉÅ (Ïòà: ÏÇ¨Ïö©Ïûê ÏûÖÎ†•Í∞í)
input_color = (120, 100, 200)  # ÏòàÎ•º Îì§Ïñ¥ Î≥¥ÎùºÏÉâ Í≥ÑÏó¥

def find_nearest_color(input_rgb, color_set):
    min_dist = float('inf')
    nearest_name = None
    for name, rgb in color_set.items():
        # Ïú†ÌÅ¥Î¶¨Îìú Í±∞Î¶¨ Í≥ÑÏÇ∞
        dist = np.sqrt(sum((a - b) ** 2 for a, b in zip(input_rgb, rgb)))
        if dist < min_dist:
            min_dist = dist
            nearest_name = name
    return nearest_name, color_set[nearest_name]

nearest_name, nearest_rgb = find_nearest_color(input_color, faber_castell_colors)

print(f"ÏûÖÎ†• ÏÉâÏÉÅ: {input_color}")
print(f"Í∞ÄÏû• Í∞ÄÍπåÏö¥ ÌååÎ≤ÑÏπ¥Ïä§ÌÖî ÏÉâÏÉÅ: {nearest_name}, RGB: {nearest_rgb}")
```

```python
# ‚ö°Ô∏è Colab ÌôòÍ≤Ω: ÌïÑÏöîÌïú ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏÑ§Ïπò
!pip install pillow numpy matplotlib

# üìÑ ÎùºÏù¥Î∏åÎü¨Î¶¨ import
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files

# üé® ÌååÎ≤ÑÏπ¥Ïä§ÌÖî ÏÉâÏó∞ÌïÑ 12ÏÉâ (RGB Í∑ºÏÇ¨Í∞í)
faber_castell_colors = {
    "White": (255, 255, 255),
    "Black": (0, 0, 0),
    "Brown": (150, 75, 0),
    "Orange": (255, 127, 0),
    "Red": (255, 0, 0),
    "Yellow": (255, 255, 0),
    "Light Green": (0, 255, 0),
    "Green": (0, 128, 0),
    "Light Blue": (0, 255, 255),
    "Blue": (0, 0, 255),
    "Purple": (128, 0, 128),
    "Pink": (255, 192, 203)
}

# üñå Í∞ÄÏû• Í∞ÄÍπåÏö¥ ÏÉâÏÉÅ Ï∞æÍ∏∞ Ìï®Ïàò
def find_nearest_color(pixel, color_set):
    min_dist = float('inf')
    nearest_rgb = (0, 0, 0)
    for rgb in color_set.values():
        dist = np.linalg.norm(np.array(pixel) - np.array(rgb))
        if dist < min_dist:
            min_dist = dist
            nearest_rgb = rgb
    return nearest_rgb

# ‚¨ÜÔ∏è ÌååÏùº ÏóÖÎ°úÎìú
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# üñº Ïù¥ÎØ∏ÏßÄ Ïó¥Í∏∞
image = Image.open(image_path).convert("RGB")
image_np = np.array(image)

# üßë‚Äçüé® Ï∂úÎ†•Ïö© Î∞∞Ïó¥ Ï¥àÍ∏∞Ìôî
output_np = np.zeros_like(image_np)

# üé® ÌîΩÏÖÄ Îã®ÏúÑÎ°ú ÏÉâÏÉÅ Î≥ÄÌôò
for i in range(image_np.shape[0]):
    for j in range(image_np.shape[1]):
        output_np[i, j] = find_nearest_color(image_np[i, j], faber_castell_colors)

# üñº Î≥ÄÌôò Í≤∞Í≥º Ïù¥ÎØ∏ÏßÄ
output_image = Image.fromarray(output_np)

# üîç ÏõêÎ≥∏Í≥º Î≥ÄÌôò Ïù¥ÎØ∏ÏßÄÎ•º Ìï®Íªò ÏãúÍ∞ÅÌôî
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(image)
axs[0].set_title("Original Image")
axs[0].axis("off")
axs[1].imshow(output_image)
axs[1].set_title("Mapped to Faber-Castell 12 Colors")
axs[1].axis("off")
plt.show()
```

```python
# ‚ö°Ô∏è Colab ÌôòÍ≤Ω: ÌïÑÏöîÌïú ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏÑ§Ïπò
!pip install pillow numpy matplotlib scikit-image

# üìÑ ÎùºÏù¥Î∏åÎü¨Î¶¨ import
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from skimage import color
from google.colab import files

# üé® ÌååÎ≤ÑÏπ¥Ïä§ÌÖî ÏÉâÏó∞ÌïÑ 12ÏÉâ (RGB Í∑ºÏÇ¨Í∞í)
faber_castell_colors = {
    "White": (255, 255, 255),
    "Black": (0, 0, 0),
    "Brown": (150, 75, 0),
    "Orange": (255, 127, 0),
    "Red": (255, 0, 0),
    "Yellow": (255, 255, 0),
    "Light Green": (0, 255, 0),
    "Green": (0, 128, 0),
    "Light Blue": (0, 255, 255),
    "Blue": (0, 0, 255),
    "Purple": (128, 0, 128),
    "Pink": (255, 192, 203)
}

# LAB ÏÉâÏÉÅÏúºÎ°ú ÎØ∏Î¶¨ Î≥ÄÌôò
faber_lab_colors = {}
for name, rgb in faber_castell_colors.items():
    # RGB ‚Üí [0,1] Î≤îÏúÑÎ°ú Ï†ïÍ∑úÌôî
    rgb_norm = np.array(rgb).reshape(1, 1, 3) / 255.0
    lab = color.rgb2lab(rgb_norm)[0, 0]
    faber_lab_colors[name] = lab

# üñå Í∞ÄÏû• Í∞ÄÍπåÏö¥ LAB ÏÉâÏÉÅ Ï∞æÍ∏∞ Ìï®Ïàò
def find_nearest_color_lab(pixel_rgb):
    # RGB ‚Üí [0,1]
    rgb_norm = np.array(pixel_rgb).reshape(1, 1, 3) / 255.0
    lab = color.rgb2lab(rgb_norm)[0, 0]
    min_dist = float('inf')
    nearest_rgb = (0, 0, 0)
    for name, lab_color in faber_lab_colors.items():
        dist = np.linalg.norm(lab - lab_color)
        if dist < min_dist:
            min_dist = dist
            nearest_rgb = faber_castell_colors[name]
    return nearest_rgb

# ‚¨ÜÔ∏è ÌååÏùº ÏóÖÎ°úÎìú
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# üñº Ïù¥ÎØ∏ÏßÄ Ïó¥Í∏∞
image = Image.open(image_path).convert("RGB")
image_np = np.array(image)

# üßë‚Äçüé® Ï∂úÎ†•Ïö© Î∞∞Ïó¥ Ï¥àÍ∏∞Ìôî
output_np = np.zeros_like(image_np)

# üé® ÌîΩÏÖÄ Îã®ÏúÑÎ°ú ÏÉâÏÉÅ Î≥ÄÌôò
for i in range(image_np.shape[0]):
    for j in range(image_np.shape[1]):
        output_np[i, j] = find_nearest_color_lab(image_np[i, j])

# üñº Î≥ÄÌôò Í≤∞Í≥º Ïù¥ÎØ∏ÏßÄ
output_image = Image.fromarray(output_np)

# üîç ÏõêÎ≥∏Í≥º Î≥ÄÌôò Ïù¥ÎØ∏ÏßÄÎ•º Ìï®Íªò ÏãúÍ∞ÅÌôî
fig, axs = plt.subplots(1, 2, figsize=(12, 6))
axs[0].imshow(image)
axs[0].set_title("Original Image")
axs[0].axis("off")
axs[1].imshow(output_image)
axs[1].set_title("Mapped to Faber-Castell (LAB-based)")
axs[1].axis("off")
plt.show()
```

```python
def factorial_recursion(n):
    if n > 1:
        return n * factorial_recursion(n-1)
    else:
        return 1

result0 = factorial_recursion(0)  # 0! = 1
result1 = factorial_recursion(1)  # 1! = 1
result2 = factorial_recursion(2)  # 2! = 2 * 1

print(result0)
print(result1)
print(result2)
```
